# 여자배구 모델별 결과 정확도

```python
df2 = pd.read_csv('모델링데이터.csv',encoding='utf8')
df2
```

```python
from sklearn.preprocessing import MinMaxScaler
data = df2.copy()
# data[['범실_평균_다섯', '디그_평균_다섯', '리시브_평균_다섯',
#        '공격종합_평균_다섯', '블로킹_평균_다섯', '세트_평균_다섯', '서브_평균_다섯', '범실_상대전적', '디그_상대전적',
#        '리시브_상대전적', '공격종합_상대전적', '블로킹_상대전적', '세트_상대전적', '서브_상대전적']] = MinMaxScaler().fit_transform(data[['범실_평균_다섯', '디그_평균_다섯', '리시브_평균_다섯',
#        '공격종합_평균_다섯', '블로킹_평균_다섯', '세트_평균_다섯', '서브_평균_다섯', '범실_상대전적', '디그_상대전적',
#        '리시브_상대전적', '공격종합_상대전적', '블로킹_상대전적', '세트_상대전적', '서브_상대전적']])
data
```

```python
# X, Y 선택
features = ['홈_어웨이', '팀코드', '상대팀코드', '범실_평균_다섯', '디그_평균_다섯', '리시브_평균_다섯',
        '블로킹_평균_다섯', '세트_평균_다섯', '서브_평균_다섯', '범실_상대전적', '디그_상대전적',
       '공격종합_상대전적', '블로킹_상대전적', '세트_상대전적', '서브_상대전적']

X = data.loc[:436,features]
y = data.loc[:436,'결과']
```

```python
# 학습, 검증 데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, y, 
#     stratify=Y,
    test_size=0.3,
    random_state=0)
```

```python
# LogisticRegression 알고리즘
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression().fit(X_train, Y_train)
lr.score(X_train, Y_train), lr.score(X_test, Y_test)
```

```python
# DecisionTree 알고리즘
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth=10).fit(X_train, Y_train)
dtc.score(X_train, Y_train), dtc.score(X_test, Y_test)
```

```python
dtc.get_depth()
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(max_depth=5).fit(X_train,Y_train)
RFC.score(X_train,Y_train), RFC.score(X_test,Y_test)
```

```python
a = ['GS칼텍스', '흥국생명', '한국도로공사', "IBK기업은행", "KGC인삼공사", "현대건설", "페퍼저축은행"]
b=list(range(0,7))
dic = dict(zip(b,a))
```

```python
for i in range(0,7):
    print(dic[i],'팀의 예측성공율 : ',RFC.score(data[data['팀코드']==i].loc[436:,features],data[data['팀코드']==i].loc[436:,'결과']))
```

```python
결과예측 = data.loc[436:,:]

결과예측['예측'] = RFC.predict(data.loc[436:,features])
결과예측['팀명']= 결과예측['팀코드'].replace(dic)
결과예측['상대팀']= 결과예측['상대팀코드'].replace(dic)
결과예측[['팀명','상대팀','결과','예측']].head(20)
```

```python
from xgboost import XGBClassifier as xgb

Xgb = xgb(learning_rate=0.04)
Xgb.fit(X_train,Y_train)
Xgb.score(X_train,Y_train), Xgb.score(X_test,Y_test)
```

```python
from lightgbm import LGBMClassifier

light = LGBMClassifier(max_depth=4)
light.fit(X_train,Y_train)
light.score(X_train,Y_train),light.score(X_test,Y_test)
```
